```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as ms

data = pd.read_csv('/home/saran.p@censanext.local/Downloads/insurance.csv')
#print(data.head())

data['sex']=data['sex'].astype('category')
data['sex']=data['sex'].cat.codes

X = data[['age','bmi']]
Y = data['sex']

x_train, x_test, y_train, y_test = train_test_split(X,data.sex,train_size=0.7)
model = LogisticRegression()
model.fit(X, Y)
y_pred=model.predict(x_test)
print(model.score(x_train,y_train))

x1 = float(input('Enter age: '))
x2 = float(input('Enter bmi: '))
user_input = pd.DataFrame({'age': [x1],'bmi':[x2]})
prediction = model.predict(user_input)
print("The predicted value is '( 1=Male / 0=Female )':", prediction[0])

sns.lmplot(x="age",y="bmi",hue='sex',data=data)
plt.show()

print("r2 score =",round(ms.r2_score(y_test,y_pred),2))
```
