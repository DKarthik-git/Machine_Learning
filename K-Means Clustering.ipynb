```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set(context="paper", palette="Blues", style="whitegrid", font_scale=1, color_codes=True)

data = pd.read_csv("C:/Users/KARTHIK/Downloads/h/Mall_Customers.csv", index_col="CustomerID")
print(data.head())

X = data.iloc[:, [2, 3]].values
print(X)

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
sns.lineplot(range(0, 1), wcss, marker="o", color="red")
plt.title("The Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(X)
y_means = kmeans.predict(X)

plt.figure(figsize=(15, 7))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_means, palette='viridis')
sns.scatterplot(
    x=kmeans.cluster_centers_[:, 0],
    y=kmeans.cluster_centers_[:, 1],
    s=200,
    color="red",
    label='Centroids',
    marker="X"
)
plt.title("Clusters of customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (1 to 100)")
plt.legend()
plt.grid(False)
plt.show()
```
